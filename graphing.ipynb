```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib_inline.backend_inline import set_matplotlib_formats
set_matplotlib_formats('retina')
plt.rcParams['figure.figsize'] = (10, 7)
plt.rcParams["errorbar.capsize"] = 5
```

```python
models = ["gpt-3.5-turbo", "sheep_duck_llama2", "OpenBuddy-70B"]
succs = [0.93,0.77,0.92]
effs = [.77,.92,.85]
scrs = [.71,.71,.78]
labels = ["Task Success", "Task Efficiency", "Score"]
succ_errs = np.array([[.08, .31, .08],[.07, .18, .05]])
effs_errs = np.array([[.20, .08, .39],[.23, .07, .15]])
scrs_errs = np.array([[.19, .31, .33],[.21, .16, .15]])
```

```python

fig, ax = plt.subplots()
xs = np.arange(1,4)
bars = ax.plot(xs-0.1,succs, '.', markersize=20)
bars = ax.plot(xs,effs, '.', markersize=20)
bars = ax.plot(xs+0.1,scrs, '.', markersize=20)

fig.tight_layout()
ax.spines['top'].set_color('#DDDDDD')
ax.spines['right'].set_color('#DDDDDD')
ax.spines['left'].set_color('#DDDDDD')
ax.spines['bottom'].set_color('#DDDDDD')
ax.tick_params(bottom=True, left=False)
ax.set_axisbelow(True)
ax.yaxis.grid(True, color='#EEEEEE')
ax.xaxis.grid(True, color='#EEEEEE')
ax.set_xticks(xs)
ax.set_xticklabels(models)
ax.legend(labels, fontsize=16)
ax.errorbar(xs-0.1, succs, yerr=succ_errs, fmt='none')
ax.errorbar(xs, effs, yerr=effs_errs, fmt='none', c='orange')
ax.errorbar(xs+0.1, scrs, yerr=scrs_errs, fmt='none', c='green')
ax.tick_params(axis='both', which='major', labelsize=16)

# ax.set_xlabel('Model', labelpad=15, color='#333333', fontsize=22)
# ax.set_title('Performance of tested models', pad=15, color='#333333',
#              weight='bold', fontsize=24)
fig.savefig("./performance_graph.pdf", format='pdf', bbox_inches='tight')
```
